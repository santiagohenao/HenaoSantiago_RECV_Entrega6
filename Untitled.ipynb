```python
import torch
from torch import nn
import numpy as np
import matplotlib.pyplot as plt

np.set_printoptions(linewidth=150)

import PIL
from glob import glob

from sklearn.metrics import f1_score

from IPython.display import display, clear_output

device = "cuda" if torch.cuda.is_available() else "cpu"
print(f"Using {device} device for torch")
```

```python
def riffle(*args):
    " riffle('ABC','abc',123') -> 'Aa1Bb2Cc3' "
    return [i for level in zip(*args) for i in level]

def TorchMLPerceptron(layer_sizes, activation_function=nn.ReLU()):
    "Builds an multilayer perceptron chaining together several linear layers, and intertwining the same activation function"
    activators = [activation_function]*len(layer_sizes)
    linear_layers = [nn.Linear(inp,out) for inp,out in zip(layer_sizes,layer_sizes[1:])]
    layers = riffle(linear_layers,activators)
    return nn.Sequential(*layers[:-1])

def train(n_epoch=5, learning_rate=0.01, weight_decay=0.01, batch_size=512):
    
    input_dim = tensors["train"]["X"].size(1) * tensors["train"]["X"].size(2)
    out_dim = 10
    hidden_dim = 50
    
    model = TorchMLPerceptron([input_dim,*[hidden_dim]*2,out_dim],activation_function=nn.Tanh())
    
    optimizer = torch.optim.Adam(model.parameters(), lr=learning_rate, weight_decay=weight_decay)
    criterion = torch.nn.CrossEntropyLoss()
    
    train_dataset = torch.utils.data.TensorDataset(tensors["train"]["X"],tensors["train"]["Y"])
    train_dataloader = torch.utils.data.DataLoader(train_dataset, batch_size=batch_size, shuffle=True)
    
    
    train_stats = {"loss":[],"f1":[],"acc":[]}
    valid_stats = {"loss":[],"f1":[],"acc":[]}
    
    fig, axes = plt.subplots(ncols=3,figsize=(12,4),facecolor='w')
    
    for epoch in range(n_epoch):
        for X_train_batch, Y_train_batch in train_dataloader:
            X_flat = torch.flatten(X_train_batch,start_dim=1)
            Y_pred = model(X_flat)
            loss = criterion(Y_pred,Y_train_batch)
            
            # backpropagation
            optimizer.zero_grad()
            loss.backward()
            optimizer.step()
            
            # train stats
            train_stats["loss"].append(loss.item())
            
            Y_pred_truncated = torch.argmax(Y_pred,1)
            train_stats["f1"].append( f1_score(Y_train_batch.detach(),Y_pred_truncated.detach(), average='macro') )
            
            acc = sum(Y_train_batch == Y_pred_truncated)/len(Y_pred)
            train_stats["acc"].append(acc)
            
            # validation stats
            model.eval()
            X_flat = torch.flatten(tensors["valid"]["X"],start_dim=1)
            Y_pred = model(X_flat)
            
            loss = criterion(Y_pred,tensors["valid"]["Y"])
            valid_stats["loss"].append(loss.item())
            
            Y_pred_truncated = torch.argmax(Y_pred,1)
            valid_stats["f1"].append( f1_score(tensors["valid"]["Y"],Y_pred_truncated.detach(), average='macro') )
            
            acc = sum(tensors["valid"]["Y"] == Y_pred_truncated)/len(Y_pred)
            valid_stats["acc"].append(acc)
            
            # plotting
            
            for i,statname in enumerate(train_stats.keys()):
                plt.sca(axes[i])
                # clean axis
                plt.cla()
                # redraw 
                plt.plot(train_stats[statname],label="train")
                plt.plot(valid_stats[statname],label="valid")
                plt.title(statname,fontsize=15)
            
            plt.tight_layout()
            handles, labels = axes[0].get_legend_handles_labels()
            fig.legend(handles, labels,bbox_to_anchor=(1.1, 1))
            clear_output(wait=True)
            display(fig)
            
    plt.close()
```

```python
files = {s:np.random.permutation(glob(f"mnist/{s}/*/*.png")) for s in ["train","valid","test"]}
```

```python
images = {}

for kind,file_list in files.items():
    Y = []
    X = []
    for file in file_list:
        Y.append(int(file.split("/")[-2]))
        X.append(np.array(PIL.Image.open(file)))
    images[kind] = {"X":X,"Y":Y}
```

```python
tensors = {}

for kind,data in images.items():
    tensor_data = {
        "X":torch.tensor(np.array(data["X"]),dtype=torch.float,device=device),
        "Y":torch.tensor(np.array(data["Y"]),dtype=torch.long, device=device)
    }
    tensors[kind] = tensor_data
```

```python
train(n_epoch=2, learning_rate=0.01, weight_decay=0.01, batch_size=512)
```
